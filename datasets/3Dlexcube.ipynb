```python
print(data.shape)
#lexcube会将a*b*c的矩阵按c*b*a来生成立方体，所以需要对256*256*64换维64*256*256
RAD_data = np.transpose(data, (2, 0, 1))
print(RAD_data.shape)
```

```python
w = lexcube.Cube3DWidget(RAD_data)#X轴对应Range，Y轴对应Angle，Z轴对应Doppler
w  #鼠标放在某一面上滚动时可以缩放对应轴，当该轴范围小于最大值时可以鼠标在该面拖动直接查看该轴为定值的切面
```

```python
w.show_sliders()#选择上述立方体的XYZ区间
```
